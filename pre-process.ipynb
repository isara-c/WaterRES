```python
from sentinelsat.sentinel import SentinelAPI, read_geojson, geojson_to_wkt
```

```python
import sys
sys.path.append('./src')

import ogr2ogr
from sentinel2_utils import get_path, kml2geojsoN, sentinel2_API, SAFE_file, object_fit
```

## 1. Set dam ID and create folder

```python
dam_id = 'BuengChawak'
```

```python
dam_path, files_path = get_path(dam_id)
```

## !! Import  ".KML" to folder "files"

## 1.1 Convet KML file to geojson and shape file

```python
kml2geojsoN(dam_id, dam_path, files_path)
```

## 2. Dowload Satellite S2 image

## connect to Copernicus Open Access Hub

```python
user = 'tleisara'
pw = 'rerone1232'
```

```python
api = SentinelAPI(user, pw, 'https://scihub.copernicus.eu/dhus')
```

### 2.1 Query image

```python
footprint = geojson_to_wkt(read_geojson('./files/' +  dam_id + '.geojson') )
```

```python
date_interval = ('20151201', '20200803')
cc = (0, 5)
```

```python
products = sentinel2_API.get_product( api, footprint, date_interval, cc )
```

- Select index of "online" product

### 2.2 Download single scene by known product id 

- Fill index that select to "product_id"

```python
product_index = 1
product_id = products.index[product_index]
```

```python
sentinel2_API.dowload_product(api, product_id)
```

## 3. Create RGB and NDWI Geotiff file from satellite image

### Extract zip from dowloaded file

```python
filename = api.get_product_odata(product_id, full = 'True')['Filename'][:-5]
SAFE_path = dam_path + '/' + filename + '.SAFE/'
```

```python
SAFE_file.extract_zip(SAFE_path)
```

### Read satellite image bands 3 and 8 

```python
dam_CRS, bands  = SAFE_file.read_bands( SAFE_path )
```

```python
SAFE_file.write_image( bands, SAFE_path )
```

## 4. Clip satellite image by bounding box of Dam

## Get Bounding Box

```python
bbox, dam_shape = object_fit.get_BBox( dam_id, files_path )
```

## Clip satellite image

```python
object_fit.crop_image( bbox, dam_shape, dam_CRS , SAFE_path )
```
